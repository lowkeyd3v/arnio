# 🚀 Arnio: High-Performance Data Quality Pipeline Tutorial

Welcome to the beginner's guide to **Arnio**! This notebook demonstrates how to build an automated, high-speed data cleaning pipeline to handle messy datasets seamlessly.

```python
import pandas as pd
import arnio as ar

# 1. Setup sample messy dataset with generic tracking tokens
raw_data = {
    "item_id": [101, 102, 102, 103, 104],
    "label": ["  alpha ", "beta", "beta", "  gamma", "delta "],
    "region": ["north", "south", "south", "NORTH", "west"]
}

df = pd.DataFrame(raw_data)
print("--- Raw Input Data Matrix ---")
print(df)
print("-" * 30)

# 2. Convert to Arnio Frame and execute High-Performance Pipeline
frame = ar.from_pandas(df)

# Run profiling signals
report = ar.profile(frame)
print("--- Arnio Profiling Summary ---")
print(report.summary())
print("-" * 30)

# Run the strict auto-clean engineering pipeline to handle spaces, casing, and duplicates
cleaned_frame = ar.auto_clean(frame, mode="strict")

# 3. Convert back to Pandas to display final dataset
processed_df = ar.to_pandas(cleaned_frame)
print("\n--- Processed Pristine Dataset via Arnio ---")
print(processed_df)
```
